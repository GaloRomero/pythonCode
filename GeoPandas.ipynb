```python
# Esto es un GeoDataFrame
# Asigno como índice del GeoDataFrame el campo 'NOMBRE' 
localidades_shp = localidades_shp.set_index("NOMBRE")
localidades_shp
```

```python
localidades_shp.head(3)
```

```python
localidades_shp.tail(3)
```

```python
# Creamos la columna 'area'
localidades_shp["area"] = localidades_shp.area
localidades_shp["area"]
```

```python
localidades_shp["perimetro"] = localidades_shp.boundary
localidades_shp["perimetro"]
```

```python
localidades_shp["centroide"] = localidades_shp.centroid
localidades_shp["centroide"]
```

```python
punto_inicial = localidades_shp["centroide"].iloc[0]
punto_inicial
```

```python
# Calculamos las distancias del punto inicial al resto de los centroides o filas (localidades) del GeoDataFrame
localidades_shp["distancia"] = localidades_shp["centroide"].distance(punto_inicial)
localidades_shp["distancia"]
```

```python
# Calculamos la distancia media de la localidad de Santa Fe (punto_inicial) al resto de localidades
localidades_shp["distancia"].mean()
```

```python
# Ploteamos las localidades ajustando la simbología en base al área de las localidades
localidades_shp.plot("area", legend=True)
```
