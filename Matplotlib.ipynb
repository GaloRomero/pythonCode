### Matplotlib

```python
# Load necessary libraries 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic Graph

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]

plt.title('Our first Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(x,y)

plt.show()
```
