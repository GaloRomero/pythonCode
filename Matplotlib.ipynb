### Matplotlib

```python
# Load necessary libraries 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label='2x', color='red', linewidth=1, marker='.', markersize=10, markeredgecolor='black')

plt.title('Our first Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4]) # change x ticks
plt.yticks([0,2,4,6,8]) # change y ticks

plt.legend()

plt.show()
```
