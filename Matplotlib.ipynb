### Matplotlib

```python
# Load necessary libraries 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Shorthand notation 
# fmt = '[color],[marker],[line]'
# plt.plot(x,y, 'r.--', label='2x')

plt.plot(x,y, label='2x', color='red', linewidth=1, linestyle= '--', marker='.', markersize=10, markeredgecolor='black')

plt.title('Our first Graph!', fontdict={'fontname': 'Arial', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4]) # change x ticks
plt.yticks([0,2,4,6,8]) # change y ticks

plt.legend()

plt.show()
```

```python
# Look for all the specific fonts in matplotlib
import matplotlib.font_manager

# Get the list of all available fonts
available_fonts = sorted([f.name for f in matplotlib.font_manager.fontManager.ttflist])
print(available_fonts)
```

#### Graph II

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph 
# figsize: This parameter defines the size of the figure in inches. It takes a tuple (width, height). 
# For example, figsize=(5, 3) sets the figure size to 5 inches in width and 3 inches in height.
# The parameter 'dpi' stands for "dots per inch" and it sets the resolution of the figure. A higher dpi results in a higher resolution image. 
# For example, dpi=300 sets the resolution to 300 dots per inch, which is high resolution and typically used for publication-quality images.
plt.figure(figsize=(5,3), dpi=300) # The total of pixels would be 15000 (5x300) x 900 (3x300)

# Shorthand notation 
# fmt = '[color],[marker],[line]'
# plt.plot(x,y, 'r.--', label='2x')

# Line number 1
plt.plot(x,y, label='2x', color='blue', linewidth=1, linestyle= '--', marker='.', markersize=7, markeredgecolor='black')

# Line number 2
# Select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# You want to show the projection of the graph phenomenom on the future
# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label= 'X^2') # Take the first 6 values: [:6]
# Plot part of the graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--') # Take the last 5 values: [5:]

# Add a title (specify font parameters with fontdict)
plt.title('Our second Graph!', fontdict={'fontname': 'Arial', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4]) # change x ticks
# plt.yticks([0,2,4,6,8]) # change y ticks

# Add a legend
plt.legend()

# Save the figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)


# Show plot
plt.show()
```

#### Bar chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels,values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```

#### Line Graph
