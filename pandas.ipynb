#### Making changes to the data

```python
df.head(5)
```

```python
# Create a total category with specific columns
# df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
#df.drop(columns=['Total'])

df['Total'] = df.iloc[:, 4:10].sum(axis=1) # axis=1 means adding horizontally / the 10th column is the one we NOT include

df.head(5)
```

```python
# Move the 'Total' column to the left of the data.frame
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]] # It is only a visual thing
df.head(5)
```

#### Filtering Data 

```python
df.loc[df['Type 1'] == 'Grass']
```

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')] # Grass and Poison
```

```python
df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')] # Grass or Poison
```

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)] # Grass and Poison and HP > 70
```

```python
# You can create a new data.frame with the filtered data

new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)] # Grass and Poison and HP > 70
new_df.head(3)
```

```python
# Reset index for the new data.frame called new_df

new_df = new_df.reset_index() # save new index as a new column

new_df
```

```python
del new_df['index'] # delete the old index column
```
